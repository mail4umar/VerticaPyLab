### Data Ingestion



The political data can be scraped from the publicly accessible website for the National Assembly of Pakistan:

```python
import requests
import pandas as pd

url = 'https://na.gov.pk/en/all_members.php'
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[-1]
```

Extracting the education data is a little more complex because the source is a PDF; the following example uses the <a href="https://pypi.org/project/tabula-py/">tabula</a> library to handle this:

```python
import tabula
```

### Feature Extraction

Start by examining the political data:

```python
p_data.head()
```

There are five columns:
- <b>col0:</b> The row index parsed by the pandas library when it read the HTML
- <b>Constituency:</b> The constituency number and city name
- <b>Name:</b> The name of the winning candidate
- <b>Party:</b> The name of the winning party
- <b>Picture:</b> The title of the image for that input

The `Picture`, `name`, and `col0` are not helpful for this analysis, so these can be dropped:

```python
p_data.drop(columns=['Picture','Name','col0'])
```

The data has some other oddities. For example, the first row only lists Khyber Pukhtunkhwa and no city or party name. Additionally, the city names are combined with the constituency identifiers (e.g. NA-1).

The final dataset should only have rows that contain a city name. Because the city names are erroneously and consistently combined with the constituency identifiers, you can use <a href="https://www.vertica.com/python/documentation_last/vdataframe/main-methods/search/">search()</a> to search for an "NA" in the `constituency` column:

```python
p_data=p_data.search(conditions='Constituency LIKE \'%NA%\'')
```

The next step is to disentangle the city names from their constituency identifiers; we need the names by themselves so they can be matched up with the education data. You can do this with [regexp()](#https://www.vertica.com/python/documentation_last/vdataframe/main-methods/regexp/). The following pattern searches the `Constituency` column for the string "NA-" followed by any number of digits, and then stores them in another column `NAs`:

```python
p_data.regexp(column = "Constituency",
            pattern = "[N]+[A]+[-]+\d{1,}",
            method = "substr",
            occurrence =1,
            name = "NAs").select(["Constituency",
                                              "NAs"])
```

Now that the constituency identifiers have been saved into another column, they can be removed from the `Constituency` column with <a href ="https://www.vertica.com/python/documentation_last/vdataframe/vcolumn-methods/str_replace/">str_replace</a>:

```python
p_data["Constituency"].str_replace(to_replace = "NA-(\d{1,})", 
                            value = "")
```

Rename the `Constituency` column to `City`:

```python
p_data.Constituency.rename('City')
```

There are multiple rows with cities that use some Roman numerals after their names (e.g. "Swat-I"). To remove these, use <a href ="https://www.vertica.com/python/documentation_last/vdataframe/vcolumn-methods/str_replace/">str_replace</a>. The `-...` pattern matches a dash ("-") followed by three characters. These calls iteratively delete the first three characters followed by a dash, then two characters, then one character:

```python
p_data["City"].str_replace(to_replace = "-...", 
                            value = "")

p_data["City"].str_replace(to_replace = "-..", 
                            value = "")

p_data["City"].str_replace(to_replace = "-.", 
                            value = "")
```

You can also identify which party got the majority seats in a particular city by finding which party is the <i>mode</i> with <a href="https://www.vertica.com/python/documentation_last/vdataframe/main-methods/analytic/">analytic()</a>. This example finds the most popular party in each city and adds it to a new column `Majority_Party`:
    
```python
p_data.analytic(func = "mode", columns = "Party", by = "City", name = "Majority_Party").groupby(["City", "Majority_Party"])
```

While this seems like it shows the most popular party in each city, there are some subtle assumptions/errors in such an analysis. For example, `analytic()` breaks ties by simply returning the first item in the list, and there could be cities where two parties won an equal number of seats.

Because a city can have multiple constituencies, it would be better to look at all the constitutencies in every city to find the relationship between political affiliation and education.

### SQL Magic Cells

Alternatively, you can more accurately and flexibly perform a join using LIKE in a SQL query. To create SQL queries, the queried tables must have a name. You can give the imported data a name when you import the data with `read_csv()`. This example specifies a table name but not a schema name, so the tables are automatically put into the default `public` schema:

```python
data=vp.read_csv('na_data.csv',table_name="party")
e_data=vp.read_csv('e_data.csv',table_name="education")
```

Another requirement is that the column names cannot contain spaces:

```python
e_data["Education Score"].rename("Education_Score")
```

To open a new cell and use a SQL query to perform the JOIN:

```python
%%sql

SELECT education.Education_Score, party.party
FROM education
INNER JOIN party ON LOWER(party.city) = LOWER(education.city);
```

When a SQL magic cell is executed, the results of the query are automatically saved in the variable `_`. The following example stores the results in `_` into another variable called `results`:

```python
results=_
```

### Analysis

Analyze the cleaned data:

```python
results.Party.value_counts(k=10)
```

To simplify the example, focus on the three main political parties (PML-N, PTI, PPPP) by filtering out the others:

```python
results.filter("Party != 'MMAP'")
results.filter("Party != 'ANP'")
results.filter("Party != 'BNP'")
results.filter("Party != 'GDA'")
results.filter("Party != 'MQMP'")
results.filter("Party != 'PML'")
results.filter("Party != 'BAP'")
results.filter("Party != 'Ind'")
```

Next, create histograms of the average `Education_Score`:

```python
results.hist(["party"], method = "avg", of = "Education_Score")
```

Based on this data, it seems that PPPP is more popular in less educated areas, whereas PML-N is more popular in more educated areas. However, the number of cities won by each party varies, so using the average in this way could skew the results. For example, a single win in an area with an education score of 80 would would result in a higher average than ten wins in areas with an education score of 75.

Visually, you can get a feeling for how varied the distribution is between parties and the populace by plotting the Gender Parity Score (or a similar metric) and the Education Score:

```python
results.scatter(["Gender Parity Score","Education Score"], 
             catcol = "Party"
               )
```

```python
pti=results.search('Party LIKE \'PTI\'')["Education Score"]
pmln=results.search('Party LIKE \'PMLN\'')["Education Score"]
```

Calculating some aggregates:
```python
pmln.aggregate(func = ["min", "25%", "50%", "75%", "max"])
```

```python
pti.aggregate(func = ["min", "25%", "50%", "75%", "max"])
```

These aggregates make the differences in education score more pronounced; the upper 25% cities in education score between the parties are almost the same, but percentiles below that have more tangible differences.

There are many other ways to find correlations between party popularity and the general populace - this is just one example to demonstrate data wrangling, analysis, and the functionalities of VerticaPy.
