# Ingestion

```python
import requests
import verticapy as vp
```

The independent variables in this example are the advertising mediums (TV, radio, and newspaper) and the dependent variable is the volume of sales.

The first step is to ingest the data:

```python
data=vp.read_csv("data/advertising.csv")
```

The next step is to review and clean the data to prepare it for analysis. The columns in the dataset are as follows:

```python
data
```

The col0 column just functions as a row count, and is therefore unnecessary. To drop this column:

```python
data.drop(columns = ["col0"])
```

Datasets often contain missing or NULL values, and these must be either removed or replaced before a model can be built. This dataset contains 200 rows, so if the dataset does not contain any missing or NULL values, the count() function should return 200 for each column:

```python
data.count()
```

This dataset does not have any NULL or missing values.

It is often easier to get an intuitive grasp of the relationships in the data with visuals, so VerticaPy contains several plotting features. The example below uses the scatter_matrix() function to create scatter plots of the data:

```python
data.scatter_matrix()
```

The next step is to analyze the correlations between the features:



```python
data.corr(method = "spearman")
```

A value greater than 0.6 indicates a strong correlation. Based on this figure and the correlation matrix, there are no strong correlations between predictors/independent variables. The strongest correlation between predictors/independent variables is between Radio and Newspaper, but at 0.317, this is only a moderate correlation.

# Model Fitting and Analysis

```python
from verticapy.learn.linear_model import LinearRegression
model = LinearRegression(name = "public.LR_winequality")
model.fit(data,["TV","Radio","Newspaper"],"Sales")
```

So far, the data suggests the following:

- The high p-value of Newspaper suggests that it does not have a strong (linear) relationship with Sales.
- TV and Radio have p-values of around 0, which suggests that they strongly affect Sales.
- The coefficient of Radio is much higher than that of TV, suggesting that a change in Radio results in a greater change in Sales than would be seen from a similar change in TV.
- The coefficients of TV and Radio are both positive, so they are positively correlated.

There are several metrics you can use to assess the accuracy of the model, such as AIC and R-squared. The regression_report() function returns a table with all such metrics:

```python
model.regression_report()
```

The combination of high adjusted R-squared (r2_adj) and low Root-mean-square error (root_mean_squared_error) confirms that the model is very accurate.

After verifying the accuracy of the model, you can use it to predict future values:



# Model modifications

You can sometimes improve the accuracy of a model by making modifications to it. One such modification is removing statistically insignificant features. In the previous example, the p-value of Newspaper was quite high, indicating a low statistical significance. To remove Newspapers and check the accuracy of the new model:



```python
model.fit(data,["TV","Radio"],"Sales")
model.regression_report()
model.score("r2adj")
```

While removing the Newspaper feature does not significantly improve the adjusted R-squared value (the accuracy) of the model, the lack of any meaningful change in R-squared confirms that Newspaper advertisements have no noticeable effect on Sales.

Another thing worth investigating is the effects of using different advertising mediums together. For example, how does advertising on both TV and Radio affect Sales? To test this, the following example adds an interaction term 
Radio x TV:



```python
data.eval(name = "TVnRadio", expr = "TV*Radio")
```

To verify how TVnRadio affects sales, calculate the p-value:



```python
model.fit(data,["TV","Radio","TVnRadio"],"Sales")
```

The TVnRadio interaction term has a very small p-value, suggesting that it has a strong effect on Sales. To verify this, check the accuracy of this new model with score():



```python
model.score("r2adj")
```

The adjusted R-squared value has jumped to 0.967, so the new model is very accurate.

Another way to potentially improve model accuracy is by including non-linear terms. For example, if you observed that the relationship between Radio and Sales is non-linear, then you could add Radio^2 as a variable.

The following example adds a Radio_2 variable to test this:



```python
data.eval(name = "Radio_2", expr = "Radio*Radio")
model.fit(data,["TV","Radio","Radio_2"],"Sales")
```

Unfortunately, the Radio_2 variable produces a high p-value, indicating that the relationship between Radio_2 and Sales is weak, so you should discard this term.

The previous examples yielded useful results and an improved model, but they still came down to trial and error based on R-squared, making it infeasible for models with many features. To overcome this, VerticaPy offers more sophisticated methods for finding the best model, such as step-wise linear regression, which adds more features and tries various combinations, and scores them based on the Bayesian Information Criterion (BIC).

The following example uses this method to generate several models with various bic values (lower is better/more accurate):



```python
from verticapy.learn.model_selection import stepwise
features=data.get_columns()
features.remove('"Sales"')
stepwise(model,
         input_relation = data, 
         X =features, 
         y = "Sales",
         direction = "forward",)
```

These results show that a combination of TV, TVnNewspaper (an interaction term), Radio, and Newspaper produces the most accurate model.
